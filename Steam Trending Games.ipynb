```python
from bs4 import BeautifulSoup
```

```python
scraped=BeautifulSoup(pagecontent,'html.parser')
steam_tab=scraped.find_all('div',{'class':'tab_content' })
```

```python
#Scraping the information of top 30 trending games
```

```python
trending_tab=scraped.find_all('div',{'id':'tab_newreleases_content'})
len(trending_tab)
```

```python
trending_content=trending_tab[0].find_all('div',{'class':'tab_item_name' })
trending_item_name=[]
for tag in trending_content:
     trending_item_name.append(tag.text)
trending_item_name[0]
```

```python
trending_a=trending_tab[0].find_all('a',{'class':'tab_item'})
trending_item_link=[]
for tag in trending_a:
     trending_item_link.append(tag['href'])
trending_item_link[0]
```

```python
trending_price=trending_tab[0].find_all('div',{'class':'discount_final_price' })
trending_item_price=[]
for tag in trending_price:
     trending_item_price.append(tag.text)
trending_item_price[0]
```

```python
trending_ge=trending_tab[0].find_all('div',{'class':'tab_item_top_tags' })
trending_item_ge=[]
for tag in trending_ge:
     trending_item_ge.append(tag.text)
trending_item_ge[0]
```

```python
#converting the scrapped information to dataframe
```

```python
dictt ={'Game':trending_item_name, 'Genre':trending_item_ge, 'price':trending_item_price, 'Link':trending_item_link}
trending_df=pd.DataFrame(dictt)
trending_df
```
